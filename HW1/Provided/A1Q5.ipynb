In this question, some basic work with MatPlotLib.  

```python
# this is the conventional import 
import numpy as np
import matplotlib.pyplot as plt
```

### Data sets
In the cell below, three data sets are generated using Numpy.  Make sure you understand what the code is doing!  

```python
# x-coordinates
xs = np.linspace(0.0, 4.0)

# 3 sets of y-coordinates
y1 = np.cos(2 * np.pi * xs) * np.exp(-xs)
y2 = np.cos(2 * np.pi * xs)
y3 = np.sin(3 * np.pi * xs)
```

### Task 1
Plot the three data sets (`xs vs y1`, `xs vs y2`, `xs vs y3`) on the same set of axes.  Make the plot look good, choosing colours, giving labels for horizontal and vertical axes, and a title.  Add a legend.  Marks will be deducted for careless presentation. 

```python
# setting bigger size than default plot size
plt.figure(figsize=(12, 4))

# plotting y1
plt.plot(xs, y1, 'k', label=r'$cos(2x)e^{-x}$')

# plotting y2
plt.plot(xs, y2, 'g', label=r'$cos(2x)$')

# plotting y3
plt.plot(xs, y3, 'y', label=r'$sin(3x)$')

# setting axis range [lowx highx lowy highy]
plt.axis([0, 4, -1, 1])

# adding x axis label
plt.xlabel('x axis')

# adding y axis abel
plt.ylabel('y axis')

# adding plot title
plt.title('Trigonometric functions plot', fontweight='bold')

# showing plot legend and adjust it outside of plot
plt.legend(bbox_to_anchor= (1.01, 1.01), ncol=1, 
            borderaxespad=0)
# showing plot, although jupyter calls this function implicitly
plt.show()
```

### Task 2
Plot the three data sets (`xs vs y1`, `xs vs y2`, `xs vs y3`) on three independent axes, **horizontally** arranged.  Make the plots look good, choosing colours, giving labels for horizontal and vertical axes, and a title.  Marks will be deducted for careless presentation.

```python
# setting bigger size than default plot size
plt.figure(figsize=(12, 4))

# the three numbers are (number of plots vertically, number of plot horizontally, which plot is this plot)
plt.subplot(131)
plt.plot(xs, y1, 'k', label=r'$cos(2x)e^{-x}$')
plt.axis([0, 4, -1, 1])
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title(r'$cos(2x)e^{-x}$')

plt.subplot(132)
plt.plot(xs, y2, 'g', label=r'$cos(2x)$')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title(r'$cos(2x)$')

plt.subplot(133)
plt.plot(xs, y3, 'y', label=r'$sin(3x)$')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title(r'$sin(3x)$')

# for making plots not to intertwine with each other
plt.tight_layout()

plt.show()
```

### Task 3
Plot the three data sets (`xs vs y1`, `xs vs y2`, `xs vs y3`) on three independent axes, **vertically** arranged.  Make the plots look good, choosing colours, giving labels for horizontal and vertical axes, and a title.  Marks will be deducted for careless presentation. 

```python
plt.figure(figsize=(12, 4))

# the three numbers are (number of plots vertically, number of plot horizontally, which plot is this plot)
plt.subplot(311)
plt.plot(xs, y1, 'r', label=r'$cos(2x)e^{-x}$')
plt.axis([0, 4, -1, 1])
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title(r'$cos(2x)e^{-x}$')

plt.subplot(312)
plt.plot(xs, y2, 'g', label=r'$cos(2x)$')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title(r'$cos(2x)$')

plt.subplot(313)
plt.plot(xs, y3, 'y', label=r'$sin(3x)$')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title(r'$sin(3x)$')

plt.tight_layout(pad=0.4, h_pad=0.25)

plt.show()
```
